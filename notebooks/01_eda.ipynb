```python
import pandas as pd
```

```python
col_names = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"
]
df = pd.read_csv("D:\\Heart_disease_predicton_pytorch\\data\\raw\processed.cleveland.data", names=col_names)

print(df.head()) # Display the first few rows of the DataFrame
```

```python
df["target"].value_counts() # Check the distribution of the target variable
```

```python
df["target"] = df["target"].apply(lambda x: 1 if x > 0 else 0) # Convert target to binary (1 for heart disease, 0 for no heart disease)
print(df["target"].value_counts()) # Check the distribution of the converted target variable
```

### 164 - heart disease free person and 
### 139 - heart disease persons.

```python
# now this should go because i want clean data
df = df.replace("?", pd.NA) # Replace "?" with NA (missing value)
df = df.dropna() # Drop rows with missing values

df.to_csv("D:\\Heart_disease_predicton_pytorch\\data\\processed\\heart_disease_data.csv", index=False) # Save the cleaned DataFrame to a new CSV file

print(df[df == "?"].count()) # Check if there are any remaining "?" values in the DataFrame (should be 0 for all columns)

#Now we have a clean dataset saved as "heart_disease_data.csv" in the "processed" folder.
```
